```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Ishan Das/OneDrive/Desktop/ZOPTAP ASSISMENT/Products.csv")

# Initial exploration of the dataset
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
# EDA - Price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True)
plt.title('Price Distribution of Products')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# EDA - Product Category distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Category', order=df['Category'].value_counts().index)
plt.title('Product Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# EDA - Average Price by Category
category_avg_price = df.groupby('Category')['Price'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
category_avg_price.plot(kind='bar', color='skyblue')
plt.title('Average Price by Product Category')
plt.xlabel('Category')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# EDA - Price vs. Product Count in Category
category_count = df['Category'].value_counts()
plt.figure(figsize=(12, 6))
sns.scatterplot(x=category_count.values, y=category_avg_price.values)
plt.title('Price vs. Product Count by Category')
plt.xlabel('Product Count')
plt.ylabel('Average Price (USD)')
plt.show()
```

```python
insights = [
    "1. Price distribution is right-skewed, with a few products priced significantly higher than the majority.",
    "2. Most products are concentrated in a few popular categories, indicating potential focus areas for marketing.",
    "3. The highest average price is found in the 'Luxury' category, which could imply premium products.",
    "4. Categories with a higher number of products tend to have lower average prices, indicating competition-driven pricing.",
    "5. Products in higher-priced categories show lower frequency of occurrence, which suggests exclusivity or niche markets."
]

# Display insights
for insight in insights:
    print(insight)
```
